```python
import scipy.stats as stats
import math
x = int(input("Sample input:"))
n = 200

p = x / n
SE = math.sqrt(p * (1 - p) / n)
alpha = 0.10
Z = stats.norm.ppf(1 - alpha / 1)

margin_of_error = Z * SE

lower_bound = p - margin_of_error
upper_bound = p + margin_of_error


print("Sample output:"f"{lower_bound:.5f} {upper_bound:.5f}",)
```

## **Promblem statement:2**

```python
import scipy.stats as stats

mean_speed = 75
std_dev = 15

X = float(input("Enter the speed X (in km/hr): "))

probability = 1 - stats.norm.cdf(X, mean_speed, std_dev)

print(f"The probability that a car is traveling faster than {X} km/hr is: {probability:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_wine
from scipy.stats import norm

wine = load_wine()

df = pd.DataFrame(wine.data, columns=wine.feature_names)



df['target'] = wine.target

df_sample = df.sample(n=50, random_state=100)

alcohol_data = df_sample['alcohol']

sample_mean = alcohol_data.mean()
sample_std = alcohol_data.std()

z_critical = norm.ppf(0.975)
n = len(alcohol_data)
margin_of_error = z_critical * (sample_std / np.sqrt(n))

confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Sample Mean (Alcohol): {sample_mean:.4f}")
print(f"Sample Standard Deviation (Alcohol): {sample_std:.4f}")
print(f"Z-Critical Value (95% Confidence): {z_critical:.4f}")
print(f"Margin of Error: {margin_of_error:.4f}")
print(f"Confidence Interval for Alcohol: ({confidence_interval[0]:.4f}, {confidence_interval[1]:.4f})")

margin_of_error = z_critical * (sample_std / np.sqrt(n))

confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Sample Mean (Alcohol): {sample_mean:.4f}")
print(f"Sample Standard Deviation (Alcohol): {sample_std:.4f}")
print(f"Z-Critical Value (95% Confidence): {z_critical:.4f}")
print(f"Margin of Error: {margin_of_error:.4f}")
print(f"Confidence Interval for Alcohol: ({confidence_interval[0]:.4f}, {confidence_interval[1]:.4f})")

```
